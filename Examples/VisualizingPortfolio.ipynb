# Portfolio Visualizer overview

Portfolio visualizer is part of *portfolio_analizer*. The labrary want to help to show in a graphical way the portfolio composition, compare your portfolio against a benchmark and so on.
This notebook aims to go througt the library to show how to use it.

### Instalation
Just installing portfolio_analizer you will be able to use it.

```python
import pandas as pd

from portfolio_visualizer.portfolio_visualizer import PortfolioVisualizer
```

## Creating a portfolio

if we don't have a portfolio, we need to create one as below

```python
portfolio = pd.DataFrame([('VTI',2),('GLD',2),('QQQ',8)],
                         columns=['ticker', 'weights'])

portfolio.head()
```

```python
# create an instance of Portfoliovisualizer
pv = PortfolioVisualizer(portfolio)
```

Portfolio visualizer allows you to save your portfolio to avoid the creation the next time. Also you can create a new instance of the library passing the portfolio file dumped.

```python
# You must provide a filepath and file name to dump the portfolio
pv.dump_portfolio('/Users/nestornav/Desktop/test_portfolio.csv')
```

```python
# If you have already saved your portfolio, you will import as below
pv_from_dump = PortfolioVisualizer.load_porfolio('/Users/nestornav/Desktop/test_portfolio.csv')
```

## Visuazalizing the portfolio

At this moment we just can show up the portfolio composition and compare against a benchmark.

```python
# Show my portfolio composition
pv.plot_pie_portfolio()
```

We have several parameter to play here. But as a naive use all of theme have default values.

### Parameters:
- from_date: initial date to fetch the prices. Default: '2000-01-01'
- to_date: Until want you want to fetch. Default: '2100-12-31',
- frecuency: Time windows to fetch the asset prices. This can be monthly, yearly and so on. Default: 'daily'
- bench: symbol ofthe benchmark to compare. Default: '^GSPC'
- allow_null: this flag allow you to fetch data from period of time where a ticker maybe didn't be created yet. This is the common case for newers companies in the market.Default: True.

```python
pv.benchmark_portfolio(allow_null=True)
```
